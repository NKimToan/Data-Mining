```python
import pandas as pd
from apyori import apriori
```

```python
df= pd.read_csv("basket_apriori.csv")
df
```

```python
df.shape
```

```python
records = []
for i in range (999):
    records.append([str(df.values[i,j]) for j in range(17)])
records
```

```python
asscociation_rules = apriori(records, min_support=0.10, min_confidence = 0.1, min_lift = 1.2, min_length = 1)
asscociation_rules = list(asscociation_rules)
print(asscociation_rules)
```
